```python
import matplotlib.pyplot as plt
```

# 1. Lineplot

```python
plt.figure(figsize=(10, 6)) # tamaño del gráfico

# pais 1
x = [2016, 2017, 2018, 2019, 2020, 2021]
y = [42, 43, 45, 47, 48, 50]

# pais 2
x2 = [2016, 2017, 2018, 2019, 2020, 2021]
y2 = [43, 43, 44, 44, 45, 45]

plt.plot(x, y, marker='o', linestyle='--', color='g', label='Colombia')
plt.plot(x2, y2, marker='d', linestyle='-', color='r', label='Argentina' )

plt.xlabel('Años')
plt.ylabel('Población (M)')
plt.title('Años vs Población')
plt.legend(loc='lower right')

plt.yticks([41, 45, 48, 51]) # solo para mostrar x números en los ejes

plt.savefig('ejemplo1.png') # para guardar la gráfica en el pc

plt.show() # solo para mostrar el gráfico simple
```

# 2. Barplot

```python
x = ['Argentina', 'Colombia', 'Perú']
y = [40, 50, 33]

plt.bar(x,y)
plt.show()
```

# 3. Piechart

```python
plt.pie(y, labels=x)
plt.show()
```

# 4. Histograma


```python
edades = [15, 16, 17, 20, 21, 21, 22, 23, 24, 25, 26, 30, 31, 32, 35, 70]
bins = [15, 20, 25, 30, 35]

plt.hist(edades, bins, edgecolor='black')
plt.show()
```

# 5. Boxplot

```python
plt.boxplot(edades)
plt.show()
```

# 6. Scatterplot

```python
a = [1, 2, 3, 4, 5, 4, 3, 2, 2, 4, 5, 6, 7]
b = [7, 2, 3, 5, 5, 7, 3, 2, 1, 4, 6, 3, 2]

plt.scatter(a, b)
plt.show()

#gráfico usado para ver relaciones entre 2 variables
```

# 7. Subplots

```python
fig, ax = plt.subplots(1, 2, sharey=True)

ax[0].plot(x, y, color='g', label='Colombia' )
ax[0].legend()

ax[1].plot(x2, y2, color='r', label='Argentina')
ax[1].legend()

plt.show()
```
