## 3.2 Agregar una nueva columna con un array

El dataframe tiene 1000 filas, por tanto, tenemos que crear un array de 1000 elementos

```python
# importar numpy
import numpy as np
```

```python
# crear un array de 1000 elementos
language_score = np.arange(0, 1000)
```

```python
# tamaño del array
len(language_score)
```

```python
# agregar una nueva columna al dataframe con un array
df_exams['language score'] = language_score
df_exams

# estamos sobreescribiendo los valores de la columna, ya que esta columna fue creada anteriormente 
```

```python
# crear números enteros aleatorios entre el 1 y 100
int_language_score = np.random.randint(1, 100, size=1000)

# el primer elemento (1) es inclusivo, el segundo elemento (100) es exclusivo
# el 1 estará incluido en el array mientras que el 100 no, el nº máx será el 99
```

```python
# el mínimo valor es inclusivo y el máximo valor es exclusivo
min(int_language_score)
max(int_language_score)
```

```python
# agregar una nueva columna al dataframe con un array
df_exams['language score'] = int_language_score
df_exams
```

```python
# eliminamos la columna repetida
df_exams.pop('language_score')
df_exams
```

```python
# crear números decimales aleatorios entre 1 y 100
np.random.uniform(1, 100, size=1000)
```

## 4.2 Operaciones en filas

```python
# calcular la suma en una fila
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# calcular el score promedio y asignar los resultados a una nueva columna
(df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'])/3
```

```python
# calcular el score promedio y asignar los resultados a una nueva columna
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'] + df_exams['language score'])/4
```
